```python
#Graph 01-Yearly growth of Victims of crime within Australia

import pandas as pd
import matplotlib.pyplot as plt
data = {'Year': ["2010", "2011", "2012","2013","2014","2015","2016","2017","2018","2019"],'Total # of Victims':[998426,1023603,1048681,980521,941780,970962,1014693,962299,965252,1026474],
       'Total population':[22172469,22522197,22928023,23297777,23640331,23984581,24389684,24773350,25171439,25464116] }
new_ds1 = pd.DataFrame.from_dict(data)

new_ds1['per 10000 ppl'] = ((new_ds1['Total # of Victims'] / new_ds1['Total population'])*10000).round(2)
# print(new_ds1)

new_ds1.plot(kind='line', x='Year',y='per 10000 ppl',ylim =(0.0,2.0), color='red',grid=True,linestyle='dashed',marker='o')
plt.ylim(340,480)
plt.title("Yearly growth of Victims of crime within Australia per 10000 people", y=1.05);
plt.ylabel("# of victims per 10000 people", labelpad=10)
Change = [0, 0.93,0.64,-7.98,-5.34,1.62,2.77,-6.63,-1.28,5.12]
index = ["2010", "2011","2012","2013","2014","2015","2016","2017","2018","2019"]
df = pd.DataFrame({'Change': Change}, index=index)
ax = df.plot.bar(rot=90,width=0.7,color='red',grid=True).legend(loc='upper center')
plt.title(" % change in crime victimisation rate", y=1.05);
# plt.ylim(0,100)
plt.ylabel("Annual % change", labelpad=14)
plt.xlabel("Year", labelpad=14)
```
